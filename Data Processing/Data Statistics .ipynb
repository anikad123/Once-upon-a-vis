Ranges/levels of quantitative and ordinal data 
- quantitative
    - max and min --> range
- ordinal
    - list of categories (unique values in column)
    - "one of many book titles such as... "

Mean and Median
- quantitative
    - average and middle functions
- ordinal 
    - positive, negative, neutral (neutral = middle)
    - not a mean or median for things like book titles
